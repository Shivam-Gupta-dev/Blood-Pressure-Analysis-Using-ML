```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import joblib

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("patient_data.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
df = df.rename(columns={'C':'Gender'})
```

```python
df.info()
```

```python
print("Before duplicates:", df.shape)

df = df.drop_duplicates()

print("After duplicates:", df.shape)
```

```python
text_cols = [
    'Gender','Age','History','Patient','TakeMedication',
    'Severity','BreathShortness','VisualChanges','NoseBleeding',
    'Whendiagnoused','ControlledDiet','Stages'
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
df['Gender'] = df['Gender'].map({
    'male':0,
    'female':1
})
```

```python
age_map = {
    '18-34':1,
    '35-50':2,
    '51-64':3,
    '65+':4
}

df['Age'] = df['Age'].map(age_map)
```

```python
binary_cols = [
    'History','Patient','TakeMedication','BreathShortness',
    'VisualChanges','NoseBleeding','ControlledDiet'
]

for col in binary_cols:
    df[col] = df[col].map({'yes':1,'no':0})
```

```python
severity_map = {
    'mild':0,
    'moderate':1,
    'severe':2
}

df['Severity'] = df['Severity'].map(severity_map)
```

```python
diagnosis_map = {
    '<1 year':1,
    '1-5 years':2,
    '5+ years':3
}

df['Whendiagnoused'] = df['Whendiagnoused'].map(diagnosis_map)
```

```python
def encode_stage(x):

    if 'normal' in x:
        return 0
    elif 'stage-1' in x:
        return 1
    elif 'stage-2' in x:
        return 2
    elif 'stage-3' in x or 'crisis' in x:
        return 3
    else:
        return np.nan

df['Stages'] = df['Stages'].apply(encode_stage)
```

```python
def convert_bp(value):

    value = str(value).strip()

    if '-' in value:
        low, high = value.split('-')
        return (float(low) + float(high)) / 2

    if '+' in value:
        return float(value.replace('+',''))

    return float(value)

df['Systolic'] = df['Systolic'].apply(convert_bp)
df['Diastolic'] = df['Diastolic'].apply(convert_bp)
```

```python
print(df.isnull().sum())

df = df.dropna()

print("After cleaning:", df.shape)
```

```python
df['Stages'] = df['Stages'].astype(int)
```

```python
scaler = MinMaxScaler()

scale_cols = ['Age','Severity','Systolic','Diastolic','Whendiagnoused']

df[scale_cols] = scaler.fit_transform(df[scale_cols])
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Stages', data=df)
plt.title("Hypertension Stage Distribution")
plt.show()
```

```python
sns.pairplot(df[['Systolic','Diastolic','Stages']], hue='Stages')
plt.show()
```

```python
X = df.drop('Stages', axis=1)
y = df['Stages']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```

```python
log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)

pred_log = log_model.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, pred_log))
print(classification_report(y_test, pred_log))
```

```python
dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

pred_dt = dt.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, pred_dt))
```

```python
rf = RandomForestClassifier()

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, pred_rf))
```

```python
svm = SVC()

svm.fit(X_train, y_train)

pred_svm = svm.predict(X_test)

print("SVM Accuracy:", accuracy_score(y_test, pred_svm))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

pred_knn = knn.predict(X_test)

print("KNN Accuracy:", accuracy_score(y_test, pred_knn))
```

```python
ridge = RidgeClassifier()

ridge.fit(X_train, y_train)

pred_ridge = ridge.predict(X_test)

print("Ridge Accuracy:", accuracy_score(y_test, pred_ridge))
```

```python
nb = GaussianNB()

nb.fit(X_train, y_train)

pred_nb = nb.predict(X_test)

print("Naive Bayes Accuracy:", accuracy_score(y_test, pred_nb))
print(classification_report(y_test, pred_nb))
```

```python
results = {
    "Logistic Regression": accuracy_score(y_test, pred_log),
    "Decision Tree": accuracy_score(y_test, pred_dt),
    "Random Forest": accuracy_score(y_test, pred_rf),
    "SVM": accuracy_score(y_test, pred_svm),
    "KNN": accuracy_score(y_test, pred_knn),
    "Ridge": accuracy_score(y_test, pred_ridge),
    "Naive Bayes": accuracy_score(y_test, pred_nb)
}

pd.DataFrame(results.items(), columns=['Model','Accuracy']).sort_values(by='Accuracy', ascending=False)
```

```python
cm = confusion_matrix(y_test, pred_log)

plt.figure(figsize=(6,5))

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.title("Confusion Matrix - Logistic Regression")

plt.show()
```
