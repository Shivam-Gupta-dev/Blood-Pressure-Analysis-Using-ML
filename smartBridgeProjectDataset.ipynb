```python
dataset['TakeMedication'] = dataset['TakeMedication'].replace({'Yes ': 'Yes'})
dataset['NoseBleeding'] = dataset['NoseBleeding'].replace({'No ': 'No'})
dataset['Systolic'] = dataset['Systolic'].replace({'121- 130': '121 - 130', '100+': '100 - 110'})

dataset['Stages'] = dataset['Stages'].replace({
    'HYPERTENSION (Stage-2)': 'HYPERTENSION (Stage-2)',
    'HYPERTENSIVE CRISI': 'HYPERTENSIVE CRISI'
})

print((dataset['Diastolic'] == '130+').sum())
print((dataset['Diastolic'] == '100+').sum())

dataset['Diastolic'] = dataset['Diastolic'].replace({'130+': '100+'})
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.to_csv('preprocessedData.csv', index=False)
```
